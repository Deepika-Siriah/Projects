```python
import pandas as pd
years = range(1881,2011)
pieces = []
columns = ['name','sex','births']
    
for year in years:
    path = 'D:/SDSU/Fall 2017/LING 572/Final project/baby_names/names/yob{0:d}.txt'.format(year)
    frame = pd.read_csv(path,names=columns)
    frame['year'] = year
    pieces.append(frame)    
    
names = pd.concat(pieces, ignore_index=True)
```

```python
import collections
names.name = names.name.astype(str).str.lower()
female_totalLetterCount = collections.Counter("".join(names.loc[names.sex.astype(str) == 'F'].name.astype(str)))
male_totalLetterCount = collections.Counter("".join(names.loc[names.sex.astype(str) == 'M'].name.astype(str)))

print("Female Letters count: ")
print(female_totalLetterCount)
print("Male Letters Count: ")
print(male_totalLetterCount)
```

```python
temp = names.loc[names.sex.astype(str) == 'F'].name.astype(str)
temp1 = []
iterator = 1
for i in temp[0:-1]:
    temp1.append(i[-1:])
female_lastLetterCount = collections.Counter("".join(temp1))
print("Female Last Letter Frequency: ")
print(female_lastLetterCount)

temp = names.loc[names.sex.astype(str) == 'M'].name.astype(str)
temp2 = []
iterator = 1
for i in temp[0:-1]:
    temp2.append(i[-1:])
male_lastLetterCount = collections.Counter("".join(temp2))
print("Male Last Letter Frequency: ")
print(male_lastLetterCount)

temp = names.name.astype(str)
temp3 = []
iterator = 1
for i in temp[0:-1]:
    temp3.append(i[-1:])
total_lastLetterCount = collections.Counter("".join(temp3))
print("Total Last Letter Frequency: ")
print(total_lastLetterCount)
```

```python
labelsF, valuesF = zip(*sorted(female_lastLetterCount.items()))
labelsM, valuesM = zip(*sorted(male_lastLetterCount.items()))
labelsT, valuesT = zip(*sorted(total_lastLetterCount.items()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
indexes = np.arange(len(labelsF))

width = 0.35 

fig, ax = plt.subplots()

rects1 = ax.plot(indexes, valuesM, color='b')
rects2 = ax.plot(indexes, valuesF, color='r')
rects3 = ax.plot(indexes, valuesT, color='g')

ax.set_title('Last Letter in the name frequency for each gender')
ax.grid(True)
ax.legend((rects1[0], rects2[0], rects3[0]), ('Men', 'Women','Total'))
ax.set_xlabel('Alphabets')
ax.set_ylabel('Frequency')
plt.xticks(indexes, labelsF)
plt.show()
plt.savefig(os.path.join('lastLetter1.png'), dpi=300, format='png', bbox_inches='tight')
```

# Last Letter frequency for females for three aggregated decades - Additionally

```python
import collections

temp = names.loc[(names.sex.astype(str) == 'F') & (names.year.astype(str) >= '1880') & (names.year.astype(str) < '1890')].name.astype(str)
import string

temp4 = []

for i in string.ascii_lowercase:
    temp4.append(i)
iterator = 1
for i in temp[0:-1]:
    temp4.append(i[-1:])
f1880_lastLetterCount = collections.Counter("".join(temp4))
print("Female Last Letter Frequency in 1880s: ")
print(f1880_lastLetterCount)

temp = names.loc[(names.sex.astype(str) == 'F') & (names.year.astype(str) >= '1940') & (names.year.astype(str) < '1950')].name.astype(str)
temp5 = []
for i in string.ascii_lowercase:
    temp5.append(i)
temp5.append
iterator = 1
for i in temp[0:-1]:
    temp5.append(i[-1:])
f1940_lastLetterCount = collections.Counter("".join(temp5))
print("Female Last Letter Frequency in 1940s: ")
print(f1940_lastLetterCount)

temp = names.loc[(names.sex.astype(str) == 'F') & (names.year.astype(str) >= '1990') & (names.year.astype(str) < '2000')].name.astype(str)
temp6 = []
for i in string.ascii_lowercase:
    temp6.append(i)
iterator = 1
for i in temp[0:-1]:
    temp6.append(i[-1:])
f1990_lastLetterCount = collections.Counter("".join(temp6))
print("Female Last Letter Frequency in 1990s: ")
print(f1990_lastLetterCount)
```

```python
sorted(f1880_lastLetterCount.items())
sorted(f1940_lastLetterCount.items())
sorted(f1990_lastLetterCount.items())
```

```python
labelsF1880, values_F1880 = zip(*sorted(f1880_lastLetterCount.items()))
labelsF1940, values_F1940 = zip(*sorted(f1940_lastLetterCount.items()))
labelsF1990, values_F1990 = zip(*sorted(f1990_lastLetterCount.items()))
valuesF1880 = []
valuesF1940 = []
valuesF1990 = []
#valuesF1880.astypes(int)
for i in range(0,26):
    valuesF1880.append((int(values_F1880[i])) -1)
    valuesF1940.append((int(values_F1940[i])) -1)
    valuesF1990.append((int(values_F1990[i])) -1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
indexes1880 = np.arange(len(labelsF1880))
indexes1940 = np.arange(len(labelsF1940))
indexes1990 = np.arange(len(labelsF1990))
indexesT = np.arange(len(labelsF1990))


width = 0.35 

fig, ax = plt.subplots()
#rects1 = ax.bar(ind, Male_count, width, color='r')

rects1 = ax.plot(indexes1880, valuesF1880, color='b')
rects2 = ax.plot(indexes1940, valuesF1940, color='r')
rects3 = ax.plot(indexes1990, valuesF1990, color='g')
#rects4 = ax.plot(indexesT, valuesT, color='y')

ax.set_title('Last Letter in the name frequency for female')
ax.grid(True)
ax.legend((rects1[0], rects2[0], rects3[0]), ('1880', '1940','1990','Total'))
ax.set_xlabel('Alphabets')
ax.set_ylabel('Frequency')
plt.xticks(indexes1990, labelsF1990)
plt.show()
plt.savefig(os.path.join('female2.png'), dpi=300, format='png', bbox_inches='tight')
```

# Analysis

The female names are not stable since we can see the trend is highly volatile across the decade plotted for the following reasons:
1. Not all the names are present in each decades.
2. We can see a rapid increase in few of the names over the succeeding decades.
3. On the other hand, few of the names (such as Mary) have been popular throughout the decades may be due to religious belief or a general trend.
4. Many names were not popular in 1880s, however it gained popularity later on.
5. Similarly, not all the letters are present in each decades.
6. We also observe that popular last letter across the decades have been some vowels like a,e,i. 
7. On the other hand, last letters such as h,n,y have been gaining popularity over the decades.
